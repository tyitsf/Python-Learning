```python
%matplotlib inline
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
from sklearn import tree
```

```python
df = pd.read_csv('data.csv')
dtr=tree.DecisionTreeClassifier(max_depth=3)
# df[['month','uProd','pProd']]
# np.array(df['month','uProd','pProd'])
dtr.fit(np.array(df[['month','uProd','pProd']]),df['Status'])
# type(df['month'])
```

```python
dot_data = \
    tree.export_graphviz(
        dtr,
        out_file = None,
        feature_names = ['month','uProd','pProd'],
        filled = True,
        impurity = False,
        rounded = True
    )
```

```python
import pydotplus
graph = pydotplus.graph_from_dot_data(dot_data)
graph.get_nodes()[7].set_fillcolor("#FFF2DD")
from IPython.display import Image
Image(graph.create_png())
```
